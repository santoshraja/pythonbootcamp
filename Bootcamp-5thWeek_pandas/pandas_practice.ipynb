```python
import numpy as np 
import pandas as pd

# Defining the three dataframes indicating the gold, silver, and bronze medal counts
# of different countries
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )

df = pd.concat([gold,silver,bronze],axis=0)
df_grpby_daymonth = df.groupby(by=['Country'])
merged_df = df_grpby_daymonth.sum()
sorted_df = merged_df.sort_values(by='Medals', ascending=False)
sorted_df
```
